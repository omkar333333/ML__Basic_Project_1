# Overfitting and Underfitting

## What is Overfitting?

**Definition:**
Overfitting happens when a machine learning model learns the training data too well, including noise and unnecessary details.

### Characteristics
- High Training Accuracy ✅
- Low Testing Accuracy ❌
- Poor performance on new data

### Example
A student memorizes all answers from previous exams but cannot solve new questions.

---

## What is Underfitting?

**Definition:**
Underfitting happens when a machine learning model does not learn enough from the training data.

### Characteristics
- Low Training Accuracy ❌
- Low Testing Accuracy ❌
- Poor performance on both training and testing data

### Example
A student studies only one chapter for the exam and cannot answer most questions.

---

## Comparison Table

| Feature | Overfitting | Underfitting |
|----------|------------|-------------|
| Learning | Too Much | Too Little |
| Training Accuracy | High | Low |
| Testing Accuracy | Low | Low |
| Model Complexity | High | Low |
| Generalization | Poor | Poor |

### 1)CLASSIFICATION MATRIX


Classification metrics help us measure how well a classification model performs.

---

# 1. Accuracy

## Definition
Accuracy is the percentage of correct predictions made by the model.

### Formula

Accuracy = (TP + TN) / (TP + TN + FP + FN)

### Example

- Total Predictions = 100
- Correct Predictions = 90

Accuracy = 90 / 100 = 90%

### Use Case

Use when classes are balanced.

---

# 2. Precision

## Definition

Precision measures how many predicted positive cases are actually positive.

### Formula

Precision = TP / (TP + FP)

### Example

- Predicted Positive = 20
- Actually Positive = 15

Precision = 15 / 20 = 75%

### Use Case

Important when False Positives are costly.

Examples:
- Spam Detection
- Fraud Detection

---

# 3. Recall

## Definition

Recall measures how many actual positive cases are correctly identified.

### Formula

Recall = TP / (TP + FN)

### Example

- Actual Positive = 25
- Correctly Predicted Positive = 20

Recall = 20 / 25 = 80%

### Use Case

Important when False Negatives are costly.

Examples:
- Disease Detection
- Cancer Prediction

---

# 4. F1-Score

## Definition

F1-Score is the balance between Precision and Recall.

### Formula

F1-Score = 2 × (Precision × Recall) / (Precision + Recall)

### Example

Precision = 75%

Recall = 80%

F1-Score = 77%

### Use Case

Used when both Precision and Recall are important.

---

# Confusion Matrix

|-------------------|-----------|-----------|

## FN (False Negative)

# Regression Metrics

Regression metrics are used to measure how well a regression model predicts numerical values.

### Examples
- House Price Prediction 🏠
- Student Score Prediction 📚
- Salary Prediction 💰

---

# 1. MAE (Mean Absolute Error)

## Definition

MAE measures the average absolute difference between actual values and predicted values.

### Formula

MAE = Σ |Actual - Predicted| / n

### Example

| Actual | Predicted |
|----------|----------|
| 90 | 85 |
| 80 | 78 |
| 70 | 75 |

Errors:

|90 - 85| = 5

|80 - 78| = 2

|70 - 75| = 5

MAE = (5 + 2 + 5) / 3

MAE = 4

### Interpretation

- Lower MAE = Better Model ✅
- Higher MAE = Worse Model ❌

---

# 2. MSE (Mean Squared Error)

## Definition

MSE measures the average squared difference between actual and predicted values.

### Formula

MSE = Σ (Actual - Predicted)² / n

### Example

| Actual | Predicted |
|----------|----------|
| 90 | 85 |
| 80 | 78 |
| 70 | 75 |

Squared Errors:

(90 - 85)² = 25

(80 - 78)² = 4

(70 - 75)² = 25

MSE = (25 + 4 + 25) / 3

MSE = 18

### Interpretation

- Lower MSE = Better Model ✅
- Penalizes large errors more heavily.

---

# 3. RMSE (Root Mean Squared Error)

## Definition

RMSE is the square root of MSE.

### Formula

RMSE = √MSE

### Example

MSE = 18

RMSE = √18

RMSE = 4.24

### Interpretation

- Lower RMSE = Better Model ✅
- Easy to understand because it is in the same unit as the target value.

---

# Comparison Table

| Metric | Formula | Best Value |
|----------|----------|----------|
| MAE | Average Absolute Error | 0 |
| MSE | Average Squared Error | 0 |
| RMSE | Square Root of MSE | 0 |

---

# Python Example

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
import numpy as np

y_true = [90, 80, 70]
y_pred = [85, 78, 75]

mae = mean_absolute_error(y_true, y_pred)
mse = mean_squared_error(y_true, y_pred)
rmse = np.sqrt(mse)

print("MAE :", mae)
print("MSE :", mse)
print("RMSE:", rmse)
```
